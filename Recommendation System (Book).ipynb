## Data Collection / Mining

```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
user = pd.read_csv(r'D:\Downloads\Users.csv')
rating = pd.read_csv(r'D:\Downloads\Ratings.csv')
book = pd.read_csv(r'D:\Downloads\Books (1).csv')
```

```python
df_final = user.merge(rating,how='left',on='User-ID')
df_final
```

```python
books = df_final.merge(book,how='left',on='ISBN')
books
```

## Data Preparation / Cleaning

```python
book1 = books.dropna()
book1
```

```python
book1.isna().sum()
```

```python
book1.shape
```

```python
book1.describe().round(2)
```

```python
book1[['city' , 'state' , 'country']] = book1['Location'].str.split(',', expand=True , n=2)
```

```python
book1
```

## Data Exploration

### Data Visualization

#### For Book Ratings

```python
plt.figure(figsize=(15,10))
sns.countplot(y='Book-Rating', data=book1,order=book1['Book-Rating'].value_counts().index[0:10])
plt.title('Book Rating')
```

#### Visualization Without 0 rating in Graph 

```python
book3 = book1[book1['Book-Rating'] != 0]
book3
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='Book-Rating', data=book3,order=book3['Book-Rating'].value_counts().index[0:10])
plt.title('Book Rating without 0')
```

```python
plt.figure(figsize=(15,6))
sns.countplot(y='Publisher', data=book1,order=book1['Publisher'].value_counts().index[0:10])
plt.title('Top 10 No of books published by a publisher')
```

```python
plt.figure(figsize=(15,15))
sns.countplot(y='Book-Title', data=book1,order=book1['Book-Title'].value_counts().index[0:10])
plt.title('Top 10 Books Read by the people')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='Book-Author', data=book1,order=book1['Book-Author'].value_counts().index[0:10])
plt.title('Top 10 No of books written by an authore')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='Year-Of-Publication', data=book1,order=book1['Year-Of-Publication'].value_counts().index[0:10])
plt.title('Year that show much of Books published')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='Location', data=book1,order=book1['Location'].value_counts().index[0:10])
plt.title('Top 10 Location where more book are taken')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='country', data=book1,order=book1['country'].value_counts().index[0:10])
plt.title('Top 10 country where more book are taken')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='state', data=book1,order=book1['state'].value_counts().index[0:10])
plt.title('Top 10 state where more book are taken')
```

```python
plt.figure(figsize=(15,10))
sns.countplot(y='city', data=book1,order=book1['city'].value_counts().index[0:10])
plt.title('city where more book are taken')
```
